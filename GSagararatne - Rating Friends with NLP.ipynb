##### Pipeline and GridSearch

```python
# Placeholders
estimators = [('normalise', MinMaxScaler()),('model', LogisticRegression())]

pipe = Pipeline(estimators)

# Grid parameters
param_grid = [{'normalise':[MinMaxScaler(), StandardScaler()],
             'model':[LogisticRegression()]},
             {'normalise':[MinMaxScaler(), StandardScaler()],
             'model':[KNeighborsClassifier()]},
             {'normalise':[None],
             'model':[DecisionTreeClassifier()]}]

# Cross val = 5
grid = GridSearchCV(pipe, param_grid, cv=5)

# Fit grid
fittedgrid = grid.fit(X_train, y_train)

# Best hyperparameters
print("Best Base Model: ", fittedgrid.best_params_)

# Best score
print("Highest cv accuracy: ", fittedgrid.best_score_)
```

#### Tuning Hyperparameters of Best Model

The hyperparameters I'll be tuning are the **C** value, **solver** and **penalty**.

```python
# Placeholders
estimators = [('normalise', MinMaxScaler()),('model', LogisticRegression())]

pipe = Pipeline(estimators)

# Grid parameters
params = [{'normalise':[MinMaxScaler(), StandardScaler()],
             'model':[LogisticRegression(random_state=42)],
             'model__C':np.arange(0.00001,10,0.5),
             'model__solver':["lbfgs","liblinear"],
             'model__penalty':['l1','l2']}]

# Cross val = 5
grid = GridSearchCV(pipe, param_grid=params, cv=5, verbose=2)

# Fit grid
fittedgrid = grid.fit(X_train, y_train)
```

```python
# Best hyperparameters
print("Best Base Model: ", fittedgrid.best_params_)

# Best score
print("Highest cv accuracy: ", fittedgrid.best_score_)
```

- Classifier: LogisticRegression
- Solver: liblinear

### Dimension Reduction

#### Principal Component Analysis (PCA)

My goal here is to produce the same accuracy of 1426 variables by using fewer variables (principal components).

Strategy:
- Scale data
- Fit and Transform PCA
- Find optimal amount of principal components

```python
# Scale data
scaler = MinMaxScaler()
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Fit PCA
pca_target = PCA(n_components = 0.52)
pca_target.fit(X_train_scaled)

# Transform
X_train_pca = pca_target.transform(X_train_scaled) 
X_test_pca = pca_target.transform(X_test_scaled)
```

```python
# Optimal PCs
pca_target.n_components_
```

From the looks of it, I only need 53 Principal Components.

```python
# Model with optimized hyperparameters
logit_pca = LogisticRegression(random_state=42, C=1.0000099999999998, solver="liblinear")

# fit
logit_pca.fit(X_train_pca, y_train)

# Evaluate
logit_pca_train_acc = logit_pca.score(X_train_pca, y_train)
logit_pca_test_acc = logit_pca.score(X_test_pca, y_test)

print("Train accuracy: ",logit_pca_train_acc)
print("Test accuracy: ",logit_pca_test_acc)
```
